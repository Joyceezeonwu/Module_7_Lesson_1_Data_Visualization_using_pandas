```python
#Import libraries
import matplotlib
from matplotlib import pyplot as plt 
import pandas as pd

#Reading the csv file
sales_data_df = pd.read_csv("company_sales_data.csv")
print(sales_data_df)
```

```python
#Comparing all the product sales data for every month using the multiple line plot. 
#Customizing lines and including marker of choice

x = sales_data_df['month_number']
y1 = sales_data_df['facecream']
plt.plot(x,y1, color = 'blue', linestyle = 'solid', marker = '.')

x = sales_data_df['month_number']
y2 = sales_data_df['toothpaste']
plt.plot(x,y2, color = 'pink', linestyle = 'dotted', marker = 'o')

x = sales_data_df['month_number']
y3 = sales_data_df['facewash']
plt.plot(x,y3, color = 'red', linestyle = 'dashed', marker = '*')

x = sales_data_df['month_number']
y4 = sales_data_df['bathingsoap']
plt.plot(x,y4, color = 'indigo', linestyle = 'dashdot', marker = 'v')

x = sales_data_df['month_number']
y5 = sales_data_df['shampoo']
plt.plot(x,y5, color = 'violet', linestyle = ':', marker = '1')

x = sales_data_df['month_number']
y6 = sales_data_df['moisturizer']
plt.plot(x,y6, color = 'orange', linestyle = '--', marker = 's')

#Putting the legend at the top left corner of the axes
plt.legend(['facecream', 'toothpaste', 'facewash', 'bathingsoap', 'shampoo', 'moisturizer'], loc = 'upper left')

#Labelling the axes properly
plt.title("JORA Group of Company")
plt.xlabel('Months')
plt.ylabel("Sales of Toiletries")

#Result display
plt.show()
```
